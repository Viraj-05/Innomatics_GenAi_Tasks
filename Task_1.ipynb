# Assignment: Mastering Prompt Templates in LangChain


```python
import time

start = time.time()

from langchain_core.prompts import PromptTemplate

print("Import time:", time.time() - start)

template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} simply"
)

start = time.time()

print(template.format(topic="AI"))

print("Format time:", time.time() - start)
```
